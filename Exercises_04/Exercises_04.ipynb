## Exercise 04.2
The exercise consist of performing a deterministic Molecular Dynamics simulation via a Lennard-Jones model of Argon. The quantities are expressed in natural units, where the fixed values $\sigma$ = 0.34 nm for length, $\epsilon / k_B$ = 120 K for temperature and $m=39.948$ amu for mass.
\
The simulations are performed in three different situations:
1. solid phase: $\rho^* = 1.1$ and $T^* = 0.8$ (cut-off radius: $r_c = 2.2$)
2. liquid phase: $\rho^* = 0.8$ and $T^* = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^* = 0.05$ and $T^* = 1.2$ (cut-off radius: $r_c = 5.0$)

By performing the experiment starting from the fcc-reticular configuration it is possible to see that in the first steps of the simulation the temperature varies a lot (it decresese sharply for the solid and the liquid and increses for the gas).
\
The temperature needs to be stabilized, in deed, in order to preform the simulation at the right conditions for the solid and the liquid the initial temperature needs to be set higher than the expected value and, while for the gas the temperature is set lower then the target one, the different behaviours will be explained at each point. In order to equilibrate the simulation, I ran the simulation without considering any of it, then looking at the output files containing the quantities one wants to study, one can notice that the simulation gets equilibrated after about $10^4$ steps, then the number of equilibration steps is set at $N_{eq} = 50000$, in order to be sure of the process realization.

1. The simulation for the solid has been performed at the initial temperature of $T_{in}^* = 1.5415$ in order to obtain the final temperature of $T_{fin}^* = (0.80033 \pm 0.00026)$ very close to the expected value of $T_{exp}^* = 0.8$.
\
The starting position is an equilibrium position for the Lennard-Jones interaction potential, so the solid system oscillates around this configuration, incresing the potential energy, then the kinetic energy decreases, so the temperature does, after a transient the system reaches an equilibrium configuration at the expected temperature, after the equilibration steps, the simulation performed gives the wantes results.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(25)

## Solid (T = 1.5415)
# Kinetic energy 
data_ekin = np.loadtxt("./dataEx2/solid/output_ekin.dat", usecols=2)
sigma_ekin = np.loadtxt("./dataEx2/solid/output_ekin.dat", usecols=3)
# Potential energy 
data_epot = np.loadtxt("./dataEx2/solid/output_epot.dat", usecols=2)
sigma_epot = np.loadtxt("./dataEx2/solid/output_epot.dat", usecols=3)
# Total energy 
data_etot = np.loadtxt("./dataEx2/solid/output_etot.dat", usecols=2)
sigma_etot = np.loadtxt("./dataEx2/solid/output_etot.dat", usecols=3)
# Pressure
data_pres = np.loadtxt("./dataEx2/solid/output_pres.dat", usecols=2)
sigma_pres = np.loadtxt("./dataEx2/solid/output_pres.dat", usecols=3)
# Temperature 
data_temp = np.loadtxt("./dataEx2/solid/output_temp.dat", usecols=2)
sigma_temp = np.loadtxt("./dataEx2/solid/output_temp.dat", usecols=3)

fig = plt.figure(figsize=[14, 12])
fig.suptitle("Solid", fontsize=24)

fig.add_subplot(321)
plt.errorbar(x, data_ekin, yerr = sigma_ekin, color="paleturquoise")
plt.xlabel("# blocks")
plt.ylabel('Kinetic Energy $[\epsilon]$')
plt.grid()

fig.add_subplot(323)
plt.errorbar(x, data_epot, yerr = sigma_epot, color="sandybrown")
plt.xlabel("# blocks")
plt.ylabel('Potential Energy $[\epsilon]$')
plt.grid()

fig.add_subplot(325)
plt.errorbar(x, data_etot, yerr = sigma_etot, color="mediumpurple")
plt.xlabel("# blocks")
plt.ylabel('Total Energy $[\epsilon]$')
plt.grid()

fig.add_subplot(322)
plt.errorbar(x, data_pres, yerr = sigma_pres, color="cornflowerblue")
plt.xlabel("# blocks")
plt.ylabel('Pressure $[\epsilon / \sigma^3]$')
plt.grid()

fig.add_subplot(324)
plt.axhline(y=0.8, color="cornflowerblue")
plt.errorbar(x, data_temp, yerr = sigma_temp, color="indianred")
plt.xlabel("# blocks")
plt.ylabel('Temperature $[\epsilon / k]$')
plt.grid()

plt.show()
```
