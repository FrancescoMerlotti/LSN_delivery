## Exercise 03.1
The exercise consist of estimating the Europian call-option price and the Europian put-optione price assuming the asset price evolves as a stochastic variable via a Geometric Brownian Motion (GBM). The constant rick-free is $r$ and the volatility is $\sigma$.
\
The main purpose of the exercise is comparing the result analytically derived by Black and Scholes of the European call and put options at time $t = 0$ and the result produce via Monte Carlo methods.

The parameters are set:
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

The chart below shows three results:
- the horizontal red line shows the analytically calculated price of a plain vanilla call option $C_{BS} = 14.975790778311286$,
- the blue graphic shows the price of call option estimated by sampling directly the final asset price at $t = T$,
- the orange graphic shows the price of the same option estimated by sampling each step of a discretized GBM with $\mu = r$ and $\sigma$.

```python
import numpy as np
import matplotlib.pyplot as plt

call_exp = 14.975790778311286

x = np.arange(100)

# Call calculated directly
data_call_dir = np.loadtxt("./dataEx1/output_call_dir.dat", usecols=2)
sigma_call_dir = np.loadtxt("./dataEx1/output_call_dir.dat", usecols=3)
# Call calculated with descrete sampling
data_call_smpl =  np.loadtxt("./dataEx1/output_call_smpl.dat", usecols=2)
sigma_call_smpl = np.loadtxt("./dataEx1/output_call_smpl.dat", usecols=3)

plt.figure(1, figsize=[8, 6])
plt.errorbar(x, data_call_dir - call_exp, yerr = sigma_call_dir, color="cornflowerblue")
plt.errorbar(x, data_call_smpl - call_exp, yerr = sigma_call_smpl, color="sandybrown")
plt.legend(["Call Price directly estimated","Call Price estimated with discrete sampling"], loc=0, fontsize=12)
plt.axhline(y = 0, color = "indianred")
plt.title("Call Option Price", fontsize=16)
plt.xlabel("Block number")
plt.ylabel("$<Price >$ - Price Calculated Analytically")
plt.grid()
plt.show()
```

As we can see the the results are consistent to each other, expecially the analytic result is in the confidence level of the two Monte Carlo estimated values.

The chart below shows three results:
- the horizontal red line shows the analytically calculated price of a plain vanilla put option $P_{BS} = 5.4595325819072364$,
- the blue graphic shows the price of put option estimated by sampling directly the final asset price at $t = T$,
- the orange graphic shows the price of the same option estimated by sampling each step of a discretized GBM with $\mu = r$ and $\sigma$.

```python
put_exp = 5.4595325819072364

x = np.arange(100)

# Put calculated directly
data_put_dir = np.loadtxt("./dataEx1/output_put_dir.dat", usecols=2)
sigma_put_dir = np.loadtxt("./dataEx1/output_put_dir.dat", usecols=3)
# Put calculated with descrete sampling
data_put_smpl = np.loadtxt("./dataEx1/output_put_smpl.dat", usecols=2)
sigma_put_smpl = np.loadtxt("./dataEx1/output_put_smpl.dat", usecols=3)

plt.figure(1, figsize=[8, 6])
plt.errorbar(x, data_put_dir - put_exp, yerr = sigma_put_dir, color="cornflowerblue")
plt.errorbar(x, data_put_smpl - put_exp, yerr = sigma_put_smpl, color="sandybrown")
plt.legend(["Put Price directly estimated","Put Price estimated with discrete sampling"], loc=0, fontsize=12)
plt.axhline(y = 0, color = "indianred")
plt.title("Put Option Price", fontsize=16)
plt.xlabel("Block number")
plt.ylabel("$<Price >$ - Price Calculated Analytically")
plt.grid()
plt.show()
```

As we can see the the results are consistent to each other, expecially the analytic result is in the confidence level of both the Monte Carlo estimated values.
